```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=3)
data_imputed = pd.DataFrame(imputer.fit_transform(df1[['Lat','Long_','Deaths']]), columns=df1[['Lat','Long_','Deaths']].columns)
```

```python
initial_data = data_imputed[['Lat', 'Long_', 'Deaths']]
initial_data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(initial_data.iloc[:,0:2],
                                                    initial_data['Deaths'],
                                                    test_size = 0.3,
                                                    random_state = 42)
```

```python
import torch.nn as nn
import torch.optim as optim
```

```python
import torch
device = torch.device('cuda' if torch.cuda.is_available else 'cpu')
```

```python
x_train_np = x_train.to_numpy().astype(np.float32)
x_test_np = x_test.to_numpy().astype(np.float32)
y_test_np = y_test.to_numpy().astype(np.float32)
y_train_np = y_train.to_numpy().astype(np.float32)
```

```python
device
```

```python
x_train_tensor = torch.tensor(x_train_np, dtype = torch.float32).to(device)
y_train_tensor = torch.tensor(y_train_np, dtype = torch.float32).to(device)
x_test_tensor = torch.tensor(x_test_np, dtype = torch.float32).to(device)
y_test_tensor = torch.tensor(y_test_np, dtype = torch.float32).to(device)
```

```python
print(x_train_tensor.isnan().sum())
print(y_train_tensor.isnan().sum())
print(x_test_tensor.isnan().sum())
print(y_test_tensor.isnan().sum())
```

```python
class ANN(nn.Module):
    def __init__(self, input, hidden1, hidden2, output):
        super(ANN, self).__init__()
        self.fc1 = nn.Linear(input, hidden1)
        self.fc2 = nn.Linear(hidden1, hidden2)
        self.fc_out = nn.Linear(hidden2, output)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.relu(self.fc1(x)) 
        x = self.relu(self.fc2(x)) 
        x = self.relu(self.fc_out(x)) 
        return x
```

```python
print(y_train_tensor.shape)
print(x_train_tensor.shape)
```

```python
input = x_train_tensor.shape[1]
hidden1 = 10
hidden2 = 7
output = 1
learning_rate = 0.01
```

```python
model = ANN(input, hidden1, hidden2, output).to(device)
print(model.parameters())
```

```python
optimizer = optim.Adam(model.parameters(), lr = learning_rate)
criterion = nn.MSELoss()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
for epoch in range(100):
    model.train()
    optimizer.zero_grad()

    outputs = model(x_train_tensor)
    loss = criterion(outputs, y_train_tensor)
    loss.backward()
    optimizer.step()

    model.eval()
    with torch.no_grad():
        predictions = model(x_test_tensor)
        predictions_np = predictions.cpu().numpy()

        r2 = r2_score(y_test_np, predictions_np)

    print(f'for epoch {epoch}/100 : loss = {loss.item()} : r2 = {r2}')
```
